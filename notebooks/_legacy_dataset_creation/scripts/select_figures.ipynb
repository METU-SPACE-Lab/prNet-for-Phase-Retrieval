```python
# list the file names in /mnt_hdd/onurcan/onurk/notebooks/example_test_result_large_small
# and print the file names
import os
from collections import defaultdict
path = "/hdd_mnt/onurcan/onurk/notebooks/example_test_result_large_small"
files = sorted(os.listdir(path))

metrics = defaultdict(dict)

for file in files:
    test_i, monte_carlo_i, *notimportant, psnr, ssim, _  = file.split("_")
    notimportant = "_".join(notimportant)
    
    metrics[(test_i, monte_carlo_i)][notimportant] = (float(psnr), float(ssim))
```

```python
initial_tta_ratios = {k: v["large_main_loop_adversarial_tta"][0] / v["small_initialization"][0] if "small_initialization" in v else float("-inf") for k, v in metrics.items()}
sorted(initial_tta_ratios.items(), key=lambda item: -item[1])
```
